```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
date = pd.date_range(start="01-01-2025", end="31-12-2025", freq="D")
# date = pd.date_range(start="01-01-2025", end="31-12-2025", freq="MS")
# date = pd.date_range(start="01-01-2025", end="31-12-2025", freq="D")
# date = pd.date_range(start="01-01-2025", end="31-12-2025", freq="D")
```

```python
ran_val = np.random.randint(500, 1000, 365)
```

```python
ran_val
```

```python
df = pd.DataFrame({"Date":date, "sales":ran_val})
```

```python
plt.plot(df["Date"],df["sales"])
```

```python
df.set_index("Date", inplace=True)
```

```python
df['sales'].resample('ME').sum()
```

```python
df.resample("W")["sales"].mean().plot(lw=2, ls="--", marker=">")
plt.title("Weekly analysis")
plt.Figure(figsize=(15,10))
plt.show()
```

```python
df["sales"].resample("W").sum() # gives weekly sum of sales
```
